```python
#making a basic bokeh graph
#importing bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data, both axis should have equal points
x=[1,2,3,4,5]
y=[6,7,4,5,3]

#prepare the output file
output_file("line.html")
#create a figure object
f=figure()
#create line plot
f.line(x,y)
#write the plot in the figure object
show(f)
```

```python
dir(f)
```

```python
help(f.circle)
```

```python
f.circle(x=[1,2,3],y=[1,2,3],size=20)
show(f)

#creating triangle
help(f.triangle)
```

```python
#creating triangle
f.triangle(x=[4,5,6],y=[4,5,6], size=[10,20,30])
show(f)
```
